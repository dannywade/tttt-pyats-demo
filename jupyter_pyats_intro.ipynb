# TTTT - pyATS and Genie

## Create and inspect a pyATS Device object

```python
from pyats.topology import Device, Testbed, Interface
from rich import inspect

device1 = Device("router1")
inspect(device1, methods=True)
```

## Create and inspect a pyATS Interface object

```python
interface1 = Interface("Eth1/1", "ethernet")
inspect(interface1, methods=True)
```

## Add IPv4 address to an Interface object

```python
interface1.ipv4 = "10.1.1.1"
inspect(interface1)
```

## Add an Interface to a Device object

```python
device1.add_interface(interface1)
inspect(device1)
```

## Create and inspect a Testbed object

```python
testbed_a = Testbed(name="firstTestbed", alias="myTestbed")
inspect(testbed_a, methods=True)
```

## Add a Device to a Testbed object

```python
testbed_a.add_device(device1)
inspect(testbed_a)
```
